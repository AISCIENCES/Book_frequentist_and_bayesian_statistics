```python
### Computing and Plotting probability mass function of a discrete random variable  
  
import numpy as np  
import matplotlib.pyplot as plt  
  
# inline magic function makes the plots to be stored in the notebook document  
%matplotlib inline   


# We repeat the experiment multiple times to get a better result of probabilities   
num_rep_exp = 100  
  
# variable num_tails_rep saves the number of tails got from each run of the experiment  
num_tails_rep = np.zeros(num_rep_exp)  
  
for i in range(num_rep_exp):  

    outcome1 = np.random.choice(['H', 'T'])  
    outcome2 = np.random.choice(['H', 'T'])  
      
    ### counting the number of tails from outcome1 and outcome2  
    if outcome1 == 'H' and outcome2 == 'H':  
        num_tails = 0  
    elif outcome1 == 'H' and outcome2 == 'T':  
        num_tails = 1  
    elif outcome1 == 'T' and outcome2 == 'H':  
        num_tails = 1  
    else:  
        num_tails = 2  
          
    # Saving the number of tails from each experiment at different indices of num_tails_rep.   
    num_tails_rep[i] = num_tails  
      
        
outcome_value, outcome_count = np.unique(num_tails_rep, return_counts=True)  
prob_count = outcome_count / len(num_tails_rep)  
  
# Now that we have tossed the coin twice for 1000 times, we plot the results  
plt.bar(outcome_value, prob_count)  
plt.ylabel("Probability")  
plt.xlabel("Outcome")  
plt.title("Probability Mass Function")
plt.show()  
```

```python
from scipy.stats import norm  
# generate random numbers from a normal probability density function (PDF) with zero mean and a standard deviation of 1: N(0,1)  
norm_pdf = norm.rvs(size=10000,loc=0,scale=1)  
plt.hist(norm_pdf, bins=100, density=1)  
plt.xlabel('Random variable value')  
plt.ylabel('Relative Frequency')  
plt.title('Probability Density Function (PDF) for a Normal Random Variable')  
plt.show() 
```

```python
from scipy.stats import bernoulli  
from matplotlib import pyplot as plt  
  
num_tosses = 1000  
# p=0.5 is for fair coin, any other value of p results in unfair coin  
fair_coin = bernoulli.rvs(p=0.5,size=num_tosses)  
plt.hist(fair_coin)  

plt.title('Bernouli Distribution for a fair coin')  
plt.xlabel('Value of Bernouli RV')  
plt.ylabel('Frequency of occurrence')  
plt.show()  

# plotting distribution for an unfair coin  
unfair_coin = bernoulli.rvs(p=0.2,size=num_tosses)  
plt.hist(unfair_coin)  
  
plt.title('Bernouli Distribution for an unfair coin')  
plt.xlabel('Value of Bernouli RV')  
plt.ylabel('Frequency of occurrence')  
plt.show()  
```

```python
## Binomial Random variable   
import numpy as np  
import seaborn as sns   
from scipy.stats import binom  
  
## n corresponds to number of trials in each experiment, size refers to total runs of the experiment, p is probability of success.   
binom_rv = binom.rvs(n=20,p=0.5,loc=0,size=100)  
  
print('Number of successes in each trial having 20 coin tosses =', binom_rv)  
  
## distplot from seaborn library is used to visualize probability distributions   
ax = sns.distplot(binom_rv, color='blue', hist_kws={"linewidth": 10,'alpha':1})    
# hist_kws specifies keywords to be used with histogram, linewidth specifies the width of bars and alpha is used to adjust the color strength  
ax.set(xlabel='Values of Binomial RV', ylabel='Relative Frequency (Probability)', title ='Binomial Distribution')  
```

```python
from scipy.stats import norm  
import matplotlib.pyplot as plt  
import numpy as np  
  
# generting a Gaussian random vaiable having 50 samples  
norm_rv1 = norm.rvs(size=50)  
print(norm_rv1)  
print('The mean of Normal RV1 is = %0.3f' % np.mean(norm_rv1))  
print('The standard deviation of Normal RV2 is = %0.3f' %np.std(norm_rv1))  
  
# plotting the probability distribution of the generated random variable  
plt.hist(norm_rv1)  
plt.title('Normal (Gaussian) Distribution')  
plt.xlabel('Value of Gaussian RV')  
plt.ylabel('Frequency of occurrence')  
plt.show()  
  
# generting a Gaussian random vaiable having 5000 samples  
norm_rv2 = norm.rvs(loc= 10, scale = 5, size=5000)  
#print(norm_rv2)  
print('The mean of Normal RV2 is = %0.3f' %np.mean(norm_rv2))  
print('The standard deviation of Normal RV2 is = %0.3f' %np.std(norm_rv2))  
  
# plotting the probability distribution of the generated random variable  
plt.hist(norm_rv2, bins=100)  
plt.title('Normal (Gaussian) Distribution')  
plt.xlabel('Value of Gaussian RV')  
plt.ylabel('Frequency of occurrence')  
plt.show()
```
